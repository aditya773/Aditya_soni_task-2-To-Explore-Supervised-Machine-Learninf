```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = df.drop("Scores",axis=1)
y = df['Scores']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.75,random_state=23)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
preds = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
rmse = np.sqrt(mean_squared_error(preds,y_test))
print(rmse)
```

```python
print(r2_score(preds,y_test))
```

```python
sns.distplot(y_test-preds)
```

```python
### we can see from the above plot that the difference b/w the predictions and test data almost matches a normal distribution.
```
